# **Parte 1: Fundamentos de bases de datos**

**Fundamentos de bases de datos y para ciencia de datos:**  

Los científicos de datos por lo general ocupan alrededor del 80% del tiempo en la preparación y limpieza de los datos para que puedan ser utilizados el modelo de Machine Learning. Para que el trabajo sea eficiente es de mucha utilidad que el conjunto de datos que trabajan sea ordenado, por ello se suele trabajar con bases de datos relacionales, ya que cumplen con esa característica.
Una base datos es una colección de información, la cual está depositada en tablas divididas en campos con un nombre asignado. En dichas tablas se tienen los registros de información que van ordenados en función de los campos que constituyen a la tabla.
Un científico de datos ocupa dicha información para ayudar a la toma de decisiones en el negocio en el que se encuentre laborando, esto con la ayuda del modelo que realice del mundo real.


**Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos:**

Un Data Warehouse es un ecosistema orientado a datos con el que se pueden realizar diversas tareas que ayudan a la toma de decisiones fundamentadas con datos provenientes de diversas fuentes y con el apoyo de análisis de datos históricos. Se componen de capas en donde la más externa es la que está expuesta al usuario por medio de un Front End para la representación visual de los datos. Una capa intermedia que está constituida por el motor de bases de datos, que es el encargado de cargarlos y almacenarlos. Por último, tenemos una capa interna que es en donde se quedan almacenados los datos.
Esta herramienta facilita la labor del científico de datos ya que le permite manipularlos sin la necesidad de tener conocimientos sólidos en algún lenguaje de programación.

# Limpieza de datos

La limpieza de datos implica observar más de cerca los problemas en los datos que ha seleccionado incluir en el análisis.

### Problema de datos 

```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv('default of credit card clients.csv', index_col=0)
df
```

```python
# Se tomarán en consideración para el análisis los datos generales de los clientes, su historial de pagos y los importes de su estado de cuenta
df = df.loc[ : , 'X1':'X17'] 
```

```python
# Se renombran algunos campos para que sea más facil identificar el tipo de información que contienen.
df.rename(columns = {'X1' : 'MontoCreditoUSD', 'X2': 'Genero', 'X3': 'Educación', 'X4': 'EstadoCivil', 'X5': 'Edad' }, inplace = True)
df
```

```python
# Verifica si hay valores nulos en alguno de los campos del dataframe
df.isnull().any()
```

### Solución: 

Se observa que en todos los campos, a excepción de "MontoCreditoUSD", existen valores faltantes. Se eliminarán los registros con valores faltantes, ya que imputarles un valor, como la media, puede contribuir al filtrado de información al momento de dividir los conjunto de entrenamiento y validación cuando se comience a trabajar en el modelo de ML. 

```python
df.dropna(inplace = True) # Se eliminan los registros con valores faltantes 
```

```python
df.isna().values.any() # Se valida si algúno de los registros del dataframe aún contiene valores faltantes
```

```python
df # El dataframe ahora tiene 36 registros menos, debido a que fueron eliminados por tener valores nulos.
```
